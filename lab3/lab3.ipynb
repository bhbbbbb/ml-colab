```python
import os
from vae.model_utils import ModelUtils, train_kmeans
from vae.model import VariationalAutoencoder, INPUT_SHAPE, D
from vae.config import Config
from vae.mode import Mode as M
from vae.dataset import DatasetMNIST, DatasetCIFAR10
from vae.preprocess import Preprocessor
```

## NMINST

```python
config = Config()
config.batch_size[M.TRAIN] = 128
config.batch_size[M.EVAL] = 1024
config.learning_rate = 1e-3
config.preprocessing = False
config.deep = False
config.capacity = 64
config.kmeans_model_path = "" ###
config.display()
```

```python
train_set = DatasetMNIST(M.TRAIN, config)
valid_set = DatasetMNIST(M.EVAL, config)
model = VariationalAutoencoder(config)
utils = ModelUtils.start_new_training(model, config)
utils.train(50, train_set, valid_set)
utils.plot_history(show=True)
```

### Visiualization

```python
utils.visualise_output(train_set)
utils.visualise_2d_lantent_space()
```

## CIFAR10

```python
config = Config()
config.batch_size[M.TRAIN] = 128
config.batch_size[M.EVAL] = 1024
config.learning_rate = 1e-5
config.preprocessing = True
config.deep = True
config.capacity = 64
config.input_shape = INPUT_SHAPE[D.CIFAR]
config.n_clusters = 32
config.kmeans_model_path = os.path.join(config.log_dir, "cifar", "cifar_32")
config.display()
```

### Train Kmeans Model

```python
train_set = DatasetCIFAR10(M.TRAIN, config)
valid_set = DatasetCIFAR10(M.EVAL, config)
train_kmeans(train_set, config)
```

### Start Training

```python
model = VariationalAutoencoder(config)
utils = ModelUtils.start_new_training(model, config)
utils.train(50, train_set, valid_set)
utils.plot_history(show=True)
```

### Visiualization

```python
utils.visualise_output(train_set)
utils.visualise_2d_lantent_space()
```
